```python
x=int(input("x:"))
temp=x
reverse=0
while temp>0:
    remainder=temp%10
    reverse=(reverse*10)+remainder
    temp//=10
if reverse==x:
    print("True")
else:
    print("False")
    
```

```python
#write a program to display a prime number in the given range  1 to 10
for num in range(1,10+1):
    if num>1:
        prime=True
        for i in range(2,num):
            if num%i==0:
                prime=False
                break
        if prime:
            print(num)
            
            
```
